## 6. Data Preparation & Train/Test Split

```python
features = ['TV', 'Radio', 'Newspaper']
X = df[features]
y = df['Sales']

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, random_state=SEED
)

print(f'Training samples : {X_train.shape[0]}')
print(f'Test samples     : {X_test.shape[0]}')
```

## 7. Cross-Validation

```python
kf = KFold(n_splits=5, shuffle=True, random_state=SEED)
cv_model = LinearRegression()
cv_scores = cross_val_score(cv_model, X, y, cv=kf, scoring='r2')

print('=== 5-Fold Cross-Validation R² Scores ===')
for i, score in enumerate(cv_scores, 1):
    print(f'  Fold {i}: {score:.4f}')
print(f'  Mean  : {cv_scores.mean():.4f}')
print(f'  Std   : {cv_scores.std():.4f}')
```

## 8. Model Training & Evaluation

```python
model = LinearRegression()
model.fit(X_train, y_train)

y_pred = model.predict(X_test)

mae  = mean_absolute_error(y_test, y_pred)
rmse = np.sqrt(mean_squared_error(y_test, y_pred))
r2   = r2_score(y_test, y_pred)

print('--- Model Performance ---')
print(f'Mean Absolute Error (MAE) : {mae:.3f}')
print(f'Root Mean Squared Error   : {rmse:.3f}')
print(f'R-squared (R² Score)      : {r2:.3f}')

print('\n--- Advertising ROI Breakdown ---')
print(f'Baseline Sales (Intercept): {model.intercept_:.3f}')
for feat, coef in zip(features, model.coef_):
    print(f'Each $1 spent on {feat:9} increases sales by: {coef:.4f}')
```

## 9. Results Visualization

```python
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# 1. Actual vs Predicted
axes[0].scatter(y_test, y_pred, alpha=0.7, color='steelblue', edgecolors='black', linewidths=0.4)
lims = [min(y_test.min(), y_pred.min()) - 1, max(y_test.max(), y_pred.max()) + 1]
axes[0].plot(lims, lims, 'r--', linewidth=1.5, label='Perfect Prediction')
axes[0].set_xlabel('Actual Sales')
axes[0].set_ylabel('Predicted Sales')
axes[0].set_title('Actual vs Predicted Sales', fontsize=13)
axes[0].legend()

# 2. Feature Coefficients
coef_series = pd.Series(model.coef_, index=features).sort_values(ascending=True)
bar_colors = ['#d73027' if c == coef_series.max() else 'steelblue' for c in coef_series]
coef_series.plot(kind='barh', ax=axes[1], color=bar_colors, edgecolor='black')
axes[1].set_title('Feature Coefficients (ROI per $1)', fontsize=13)
axes[1].set_xlabel('Coefficient Value')
for i, v in enumerate(coef_series):
    axes[1].text(v + 0.002, i, f'{v:.4f}', va='center', fontsize=9)

plt.tight_layout()
plt.savefig('model_results.png', dpi=150, bbox_inches='tight')
plt.show()
print('Results plots saved.')
```

## 10. Residual Analysis

```python
residuals = y_test - y_pred

fig, axes = plt.subplots(1, 2, figsize=(13, 4))

# Residuals vs Fitted
axes[0].scatter(y_pred, residuals, alpha=0.7, color='seagreen', edgecolors='black', linewidths=0.4)
axes[0].axhline(0, color='red', linestyle='--', linewidth=1.5)
axes[0].set_xlabel('Fitted Values')
axes[0].set_ylabel('Residuals')
axes[0].set_title('Residuals vs Fitted', fontsize=13)

# Residual Distribution
axes[1].hist(residuals, bins=15, color='seagreen', edgecolor='black', alpha=0.8)
axes[1].axvline(0, color='red', linestyle='--', linewidth=1.5)
axes[1].set_xlabel('Residual')
axes[1].set_ylabel('Frequency')
axes[1].set_title('Residual Distribution', fontsize=13)

plt.tight_layout()
plt.savefig('residual_analysis.png', dpi=150, bbox_inches='tight')
plt.show()
print('Residual plots saved.')
```
